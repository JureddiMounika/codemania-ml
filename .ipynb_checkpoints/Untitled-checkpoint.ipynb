```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x2 = np.arange(-10, 10, 1)
x = np.arange(0.001, 0.999, 0.0001)
def logit(p):
    return np.log((p)/(1-p))
y = []
for i in x:
    y.append(logit(i))
plt.subplot(1,2,1)
plt.plot(x,y)

def sigmoid(x):
    return (1/(1+np.exp(-x)))
y2 = []
for i in x2:
    y2.append(sigmoid(i))
plt.subplot(1,2,2)
plt.plot(x2, y2)



```

```python
from sklearn.datasets import load_breast_cancer
% matplotlib inline
from matplotlib import pyplot as plt
dataset = load_breast_cancer()
#print(dataset.DESCR)
import numpy as np
data = dataset.data
y = dataset.target
#print(y)
# print(len(data))
# print(len(data[:,0]))
# plt.plot(data[:,0], y, 'bo')
x = []
for target in range(2):
    x.append([[], []])
    for i in range(len(data)):
        if dataset.target[i] == target:
            x[target][0].append(data[i][0])
            x[target][1].append(data[i][1])
colors = ('r', 'b')
fig = plt.figure(figsize = (10, 8))
ax = fig.add_subplot(111)
for target in range(2):
    ax.scatter(x[target][0], x[target][1], c = colors[target])









```

```python
from sklearn.linear_model import LogisticRegression
log_regress = LogisticRegression()
mean_radius = data[:,0]
x = np.reshape(mean_radius, (len(mean_radius), 1))


log_regress.fit(x, y)
print(log_regress.predict([[20]]))
print(log_regress.predict_proba([[20]]))

# print(log_regress.coef_)
# print(log_regress.intercept_)
x1 = np.arange(0,30,0.01)
def sigmoid2(x):
    a = log_regress.coef_[0][0]
    b = log_regress.intercept_[0]
    e = np.exp(-(a*x + b) )
    return (1/(1+e))
y1 = [sigmoid2(i) for i in x1]
# for i in x1:
#     y1.append(sigmoid2(i))

plt.plot(x1,y1)
plt.plot(mean_radius, y, 'ro')



```
